```python
#Desafio Versão 1.0

menu = """

[d] Depositar
[s] Sacar
[e] Extrato
[q] Sair

=> """
saldo = 0
LIMITE = 500
extrato = ' '
numero_saques = 0
LIMITE_SAQUES = 3

while True:
    opcao = input(menu)
    # Deposito
    if opcao == 'd':
        valor_deposito = int(input('Quanto deseja depositar?\n'))
        saldo += valor_deposito
        print('Seu depósito foi realizado com sucesso!')
        extrato+f"Você realizou um depósito de R$ {saldo:.2f} reais.\n"
    
    # Saque
    elif opcao == 's':
        valor_sacado = int(input('Quanto deseja sacar?\n'))
        if numero_saques > LIMITE_SAQUES:
            print('Você não pode mais sacar hoje. Seu limite foi atingido!!')
        elif valor_sacado > saldo:
            print('Não é possível sacar por falta de saldo!')
        elif valor_sacado > LIMITE:
            print('O limite de saque é de R$ 500.00 reais!')
        else:
            numero_saques+=1
            saldo -= valor_sacado
            extrato.append(f'Você realizou um saque de R$ {valor_sacado:.2f} reais.')
            print('Seu saquei foi realizado com sucesso!')
            
    # Extrato
    elif opcao == 'e':
        if len(extrato) == 0:
            print('Não foram realizadas movimentações!')
        else:
            print(' EXTRATO '.center(50,'-'))
            for i in extrato: print(i)
            print(f'Seu saldo atual é de {saldo:.2f}')

    elif opcao == 'q':
        break
    else:
        print('Operação inválida! Por favor selecione a opção desejada.')
```

```python
# Desafio Versão 2.0
import textwrap 

def menu():
    menu = f"""
    {'MENU'.center(50,'=')}
[d]\tDepositar
[s]\tSacar
[e]\tExtrato
[nc]\tNovo Cliente
[cc]\tCriar Conta
[lc]\tListar Contas
[q]\tSair

=> """
    return input(textwrap.dedent(menu))

def depositar(saldo, valor, extrato,/):
    if valor > 0:
        saldo += valor
        extrato += f"Depósito:\t\t R$: {valor:.2f}\n"
        print('\n=== Seu depósito foi realizado com sucesso! ===')
    else:
        print('Valor informado inválido! Por favor, tente novamente!')

    return saldo, extrato

def sacar(*,saldo, valor, extrato, limite, numero_saques, LIMITE_SAQUES):
    if numero_saques > LIMITE_SAQUES:
        print('Você não pode mais sacar hoje. Seu limite de saques foi atingido!!')
    elif valor > saldo:
        print('Não é possível sacar! Saldo insuficiente!')
    elif valor > LIMITE:
        print('O limite de saque é de R$ 500.00 reais!')
    else:
        numero_saques+=1
        saldo -= valor
        extrato += f"Saque:\t\t R$: {valor:.2f}\n"
        print('Saque realizado com sucesso!')

    return saldo, extrato

def exibir_extrato(saldo,/,*, extrato):
    if len(extrato) == 0:
        print('Não foram realizadas movimentações!')
    else:
        print(' EXTRATO '.center(50,'-')+'\n\ extrato \n\n'+f'Seu saldo atual é de {saldo:.2f}')

def filtrar_usuario(cpf,usuarios):
    usuarios_filtrados = [usuario for usuario in usuarios if usuario['cpf'] == cpf]
    return usuarios_filtrados[0] if usuarios_filtrados else None

def criar_usuario(usuarios):
    cpf = input('Informe o CPF (somente número): ')
    usuario = filtrar_usuario(cpf,usuarios)

    if usuario:
        print('Já existe um usuário associado à este CPF!')
        return 
    
    nome = input('Informe o nome completo: ')
    data_nascimento = input('Informe a data de nascimento (dd-mm-aaaa): ')
    endereco = input('Informe o endereço (logradouro, n° - bairro - cidade/sigla do estado): ')

    usuarios.append({
        "nome": nome,
        "data_nascimento": data_nascimento,
        'cpf': cpf,
        'endereco': endereco
    })
    print('Usuário cadastrado com sucesso!')

def criar_conta(agencia, numero_conta, usuarios):
    cpf = input('Informe o CPF do usuário: ')
    usuario = filtrar_usuario(cpf, usuarios)

    if usuario:
        print('Conta criada com sucesso!')
        return {
            'agencia': agencia,
            'numero_conta': numero_conta,
            'usuario': usuario
        }

    print('Usuário não encontrado! Por favor, tente novamente!')

def listar_contas(contas):
    for conta in contas:
        linha = f"""\
            Agência:\t{conta['agencia']}
            C/C:\t\t{conta['numero_conta']}
            Titular:\t{conta['usuario']['nome']}
        """
        print('=' * 100)
        print(textwrap.dedent(linha))

def main():
    LIMITE = 500
    LIMITE_SAQUES = 3
    AGENCIA = '0001'

    numero_conta = 0
    usuarios = list()
    contas = list()
    saldo = 0
    extrato = ' '
    numero_saques = 0

    while True:
        opcao = menu()

        # Deposito
        
        if opcao == 'd':
            valor = float(input('Quanto deseja depositar?\n'))
            saldo, extrato = depositar(saldo,valor, extrato)
        
        # Saque
        
        elif opcao == 's':
            valor = float(input('Quanto deseja sacar?\n'))
            saldo, extrato = sacar(
                saldo = saldo, 
                valor = valor, 
                extrato = extrato, 
                limite = LIMITE, 
                numero_saques = numero_saques, 
                LIMITE_SAQUES = LIMITE_SAQUES
            )                    
        
        # Extrato
        
        elif opcao == 'e':
            exibir_extrato(saldo, extrato=extrato)

        # Criando usuário
        elif opcao == 'nc':
            criar_usuario(usuarios)

        elif opcao == 'cc':
            conta = criar_conta(AGENCIA, numero_conta, usuarios)
            
            if conta:
                contas.append(conta)
                numero_conta +=1
        elif opcao == 'lc':
            listar_contas(contas)
        elif opcao == 'q':
            break
        else:
            print('Operação inválida! Por favor selecione a opção desejada.')

main()
```
